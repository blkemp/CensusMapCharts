```python
# Import statements
# Declare Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import make_scorer, r2_score, mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
import os
import pickle
from textwrap import wrap
import operator
import au_census_analysis_functions as cnss_func
import geopandas as gpd
import plotly.express as px
from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable
from matplotlib.colors import TwoSlopeNorm

# Set a variable for current notebook's path for various loading/saving mechanisms
nb_path = os.getcwd()
```

```python
# Read SHP file
SA2_gdf = gpd.read_file('{}/Shapefiles/SA2/SA2_2021_AUST_GDA2020.shp'.format(nb_path))    # Load the data using Geopandas
```

```python
# Test the import worked
SA2_gdf.rename(columns={'SA2_CODE21' : 'Code'}) #this renaming isn't working?
SA2_gdf.head()
```

```python
# Import the G01 table which has basic population data for 2021 and 2016 
df_2021 = cnss_func.load_census_csv(table_list=['G01'],statistical_area_code='SA2', year=2021).iloc[:,[0,3]]
df_2021.columns=['Code','TotalPop']

df_2016 = cnss_func.load_census_csv(table_list=['G01'],statistical_area_code='SA2', year=2016).iloc[:,[0,3]]
df_2016.columns=['Code','TotalPop']
df_2021.head()
```

```python
# merge the dataframes and create a "Population change" feature
df_pops = df_2016.merge(df_2021,left_on='Code', right_on='Code')
df_pops.columns=['Code','2016_Population','2021_Population']
df_pops['Change_2016_to_2021'] = (df_pops['2021_Population']-df_pops['2016_Population'])

# create change % feature
df_pops['Change_pc'] = ((df_pops['2021_Population']/df_pops['2016_Population'])-1)*100

df_pops.head()
```

```python
# convert code to Str for merging with geopandas table
df_pops['Code'] = df_pops['Code'].astype('str')

# merge
df_merged = pd.merge(SA2_gdf[['SA2_CODE21','SA2_NAME21','geometry']], 
                     df_pops[['Code','Change_2016_to_2021', 'Change_pc']],
                     left_on='SA2_CODE21',
                     right_on='Code')
df_merged.head()
```

```python
#Checking out the out-migration
df_merged.sort_values(by = 'Change_2016_to_2021', ascending = True).head(10)
```

```python
#Checking out the in-migration
df_merged.sort_values(by = 'Change_2016_to_2021', ascending = False).head(10)
```

```python
# Get a view on maximum percentage change excluding SA2s that had no pop in 2016
df_merged[df_merged['Change_pc']!=np.inf].sort_values(by = 'Change_pc', ascending = False).head(10)
```

```python
# convert the table to relevant spatial data format and json
df_merged = df_merged.to_crs(epsg=4327)
sa2_json = df_merged.__geo_interface__
```

```python
# Plot a static choropleth using matplotlib functionality. Note because of the size you need to click on the image to zoom in.
tdf = df_merged.copy()

fig, ax = plt.subplots(1, figsize=(50,20))

# make the color bar be centered at zero
norm = TwoSlopeNorm(vcenter=0, vmin = tdf.Change_2016_to_2021.min(), vmax = tdf.Change_2016_to_2021.max())

# legend scaling so it is the same size as the map
divider = make_axes_locatable(ax)
cax = divider.append_axes("bottom", size="5%", pad=0.1)

# Plot 
tdf.plot(column='Change_2016_to_2021', cmap='RdYlGn',
         ax=ax, cax=cax, norm = norm, legend=True,
         legend_kwds={'label': "Pop Change", 'orientation' : 'horizontal'},
         edgecolor='black', linewidth=0.5)

# Plot Aesthetics
ax.set_title('Pop Change 2016 to 2021')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
plt.tight_layout()
plt.show()
```

```python
# Bring in descriptions for geographies to allow filtering by state/SA3/SA4
df_geog_ref = pd.read_excel('{}/Metadata/2016Census_geog_desc_1st_2nd_3rd_release.xlsx'.format(nb_path), sheet_name = '2016_ASGS_Main_Structures')
df_geog_ref.head()
```

```python
# compile reference columns from code
df_merged['State_Code'] = df_merged['SA2_CODE21'].str[:1]
df_merged['SA4_Code'] = df_merged['SA2_CODE21'].str[:3]
df_merged['SA3_Code'] = df_merged['SA2_CODE21'].str[:3]

# change excel file code references to string type to align to geopandas df
df_geog_ref['AGSS_Code_2016'] = df_geog_ref['AGSS_Code_2016'].astype('str')

# add in state, SA4 and SA3 descriptions
df_merged['State'] = pd.merge(df_merged[['State_Code']], 
                     df_geog_ref[['AGSS_Code_2016','Census_Name_2016']],
                     left_on='State_Code',
                     right_on='AGSS_Code_2016')['Census_Name_2016']

df_merged['SA4'] = pd.merge(df_merged[['SA4_Code']], 
                     df_geog_ref[['AGSS_Code_2016','Census_Name_2016']],
                     left_on='SA4_Code',
                     right_on='AGSS_Code_2016')['Census_Name_2016']

df_merged['SA3'] = pd.merge(df_merged[['SA3_Code']], 
                     df_geog_ref[['AGSS_Code_2016','Census_Name_2016']],
                     left_on='SA3_Code',
                     right_on='AGSS_Code_2016')['Census_Name_2016']
```

```python
# Turn matplotlib chart into a function
def plot_census_map(dataframe, plot_field, plot_title = None):
    ''' Plot a static choropleth using matplotlib functionality.
    
    INPUTS
    dataframe: GEOPANDAS DATAFRAME - the dataframe to use as a source for the plot, 
                derived from a shapefile and carrying the datafield you wish to graph
    plot_field: STRING - the column name from the dataframe you wish to use for colour intensity
    plot_title: STRING (optional) - the description of the plot you wish to use
    
    '''
    if plot_title is None: plot_title = plot_field
    
    tdf = dataframe.copy()

    fig, ax = plt.subplots(1, figsize=(50,20))

    # make the color bar be centered at zero
    # for min/max on data scale, filter out infinities then outliers based on 3 sigma variation
    ref_series = tdf[tdf[plot_field]!=np.inf][plot_field].dropna()
    color_min = max(ref_series.mean()-(3 * ref_series.std()), ref_series.min())
    color_max = min(ref_series.mean()+(3 * ref_series.std()), ref_series.max())
    
    norm = TwoSlopeNorm(vcenter=0, vmin = color_min, vmax = color_max)

    # legend scaling so it is the same size as the map
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("bottom", size="5%", pad=0.1)

    # Plot 
    tdf.plot(column=plot_field, cmap='RdYlGn',
             ax=ax, cax=cax, norm = norm, legend=True,
             legend_kwds={'label': plot_title, 'orientation' : 'horizontal'},
             edgecolor='black', linewidth=0.5)

    # Plot Aesthetics
    ax.set_title(plot_title)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.axes.xaxis.set_visible(False)
    ax.axes.yaxis.set_visible(False)
    plt.tight_layout()
    plt.show()

plot_census_map(df_merged, 'Change_pc', 'Population Change (%) - 2016 to 2020')
```

```python
# Produce map for Victoria
df_vic = df_merged[df_merged['State'] == 'Victoria']
plot_census_map(df_vic, 'Change_pc', 'Population Change (%) - 2016 to 2020')
```

```python
# Produce map for Victoria
df_vic = df_merged[df_merged['State'] == 'Victoria']
plot_census_map(df_vic, 'Change_2016_to_2021', 'Population Change (abs) - 2016 to 2020')
```
